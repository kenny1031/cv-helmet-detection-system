```python
import fiftyone as fo
import numpy as np
from collections import Counter
```

```python
dataset = fo.load_dataset("helmet_hardhats_csv")
```

```python
bbox_counts = []
bbox_areas = []

for sample in dataset:
    dets = sample["ground_truth"].detections
    bbox_counts.append(len(dets))
    for d in dets:
        w = d.bounding_box[2]
        h = d.bounding_box[3]
        bbox_areas.append(w * h)

print("Avg bboxes per image:", np.mean(bbox_counts))
print("Max bboxes in image:", np.max(bbox_counts))

print("Avg bbox area:", np.mean(bbox_areas))
print("Min bbox area:", np.min(bbox_areas))
print("Max bbox area:", np.max(bbox_areas))
```
